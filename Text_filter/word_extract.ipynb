```python
import cv2,numpy as np,pandas as pd,os,cupy as cp
from sklearn.cluster import KMeans as KM
```

```python
def does(x):
    pass

    
def coordi(items,iceri):
    max_x=0
    min_x=0
    max_y=0
    min_y=0
    for things in items:
        
        things=things[0]
        min_x=l=things[0]
        min_y=m=things[1]
        if (l>max_x):
            max_x=l
        if(l<min_x):
            min_x=l
        if (m>max_y):
            max_y=m
        if(m<min_y):
            min_y=m    
    return((min_x-iceri,min_y-iceri),(max_x+iceri,max_y+iceri))      

def mapper(contors):
     dicti={}
     for it in contors:
        
        l,n,m=it.shape
        if(l>10 and l<90):
            print("GGGGGGGGGGGGGGGGGGGGGGGGGGGGGGGGGGGGGGGGGGGGGGGGGGGGGGGGGGGGGGGGGGGGGGGG")
            cnt = it
            M = cv2.moments(cnt)
            if (M['m00'] != 0):
                cx = int(M['m10']/M['m00'])
                dicti[cx]=it
            
     return dicti        
```

```python
x=5
blanl_loc="temp//blank.png"
loc="images"
im,si=(800,800)
cv2.namedWindow("clusters")
cv2.createTrackbar("cluster","clusters",7,200,does)
for files in os.listdir(loc): 
    print(files)
    path=loc+"//"+files
    print(path)
    img=cv2.imread(path)
    # imgs=img=cv2.resize(img,(im,si))
    imgs=img.copy()
    im,si,_=imgs.shape
    blank=(np.ones((im,si,3))*255).astype(np.uint8)
    cv2.imwrite(blanl_loc,blank)
    cv2.imshow("original",imgs)
    img_grey=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
    _,bin=cv2.threshold(img_grey,127,255,cv2.THRESH_BINARY)
    img=img/255
    img_clus=cp.reshape(img,(-1,3))
    km=KM(n_clusters=x).fit(img_clus)
    new_img=km.cluster_centers_[km.labels_]
    new_img=(new_img*255).astype(np.uint8)
    new_img=cp.reshape(new_img,(im,si,3))
    cv2.imshow("clusters",new_img)
    x=cv2.getTrackbarPos("cluster","clusters")
    img_grey=cv2.cvtColor(new_img,cv2.COLOR_RGB2GRAY)
    _,bin=cv2.threshold(img_grey,127,255,cv2.THRESH_BINARY)
    # canny=cv2.Canny(img_grey,10,200)
    # cv2.imshow("canny",canny)
    ker_5x5=np.ones((1,1)).astype(np.uint8)
    # cv2.imshow("window",bin)
    bin=cv2.bilateralFilter(bin,d=5,sigmaColor=12,sigmaSpace=12)
     
    cv2.imshow("bin",bin)
    # bin_used=bin.astype(np.uint32)
    c,h=cv2.findContours(bin,cv2.RETR_CCOMP,cv2.CHAIN_APPROX_SIMPLE)
    counter_img=cv2.drawContours(blank,c,-1,(0,0,255),2)
    dicti=mapper(c) 
    print(dicti)
    keys=sorted(dicti)
    for it in keys:
        it=dicti[it] 
        blank=cv2.imread(blanl_loc)
        p,y=coordi(it,9)
        y0,y1=p[1],y[1]
        x0,x1=p[0],y[0]
        blank= cv2.rectangle(blank,p,y,(0,255,0),2,1)
        blank[y0:y1,x0:x1]=imgs[y0:y1,x0:x1]
        cv2.imshow(blanl_loc,blank)
        cv2.imwrite(blanl_loc,blank)
        cv2.waitKey(1)
    if cv2.waitKey(1) &0xff==ord("q"):
        break
    cv2.waitKey(0)
    
cv2.destroyAllWindows()
```
